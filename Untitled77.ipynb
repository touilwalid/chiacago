```python
import pandas as pd
```

```python
path = '/content/drive/MyDrive/CodingDojo/04-DataViz/Week13/Data/chiacago.csv'
df = pd.read_csv(path)
df.head()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Hour'] = df['Date'].dt.hour
am_rush_hour = df[(df['Hour'] >= 7) & (df['Hour'] < 10)]
pm_rush_hour = df[(df['Hour'] >= 16) & (df['Hour'] < 19)]
total_am_rush_hour = len(am_rush_hour)
total_pm_rush_hour = len(pm_rush_hour)
print(f"Total number of crimes during AM rush hour: {total_am_rush_hour}")
print(f"Total number of crimes during PM rush hour: {total_pm_rush_hour}")
top_5_am_crimes = am_rush_hour['Primary Type'].value_counts().head(5)
print("Top 5 most common crimes during AM rush hour:")
print(top_5_am_crimes)
top_5_pm_crimes = pm_rush_hour['Primary Type'].value_counts().head(5)
print("Top 5 most common crimes during PM rush hour:")
print(top_5_pm_crimes)
am_motor_vehicle_thefts = am_rush_hour[am_rush_hour['Primary Type'] == 'MOTOR VEHICLE THEFT']
pm_motor_vehicle_thefts = pm_rush_hour[pm_rush_hour['Primary Type'] == 'MOTOR VEHICLE THEFT']
print(f"Motor Vehicle Thefts during AM rush hour: {len(am_motor_vehicle_thefts)}")
print(f"Motor Vehicle Thefts during PM rush hour: {len(pm_motor_vehicle_thefts)}")

if len(am_motor_vehicle_thefts) > len(pm_motor_vehicle_thefts):
    print("Motor Vehicle Thefts are more common during AM rush hour.")
else:
    print("Motor Vehicle Thefts are more common during PM rush hour.")
```

```python
import matplotlib.pyplot as plt
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Month'] = df['Date'].dt.month
total_crimes_per_month = df['ID'].groupby(df['Month']).count()
plt.figure(figsize=(10, 6))
total_crimes_per_month.plot(kind='bar')
plt.title('Total Number of Crimes per Month')
plt.xlabel('Month')
plt.ylabel('Number of Crimes')
plt.xticks(range(12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
plt.grid(True)
plt.show()
```

```python
most_crimes_month = total_crimes_per_month.idxmax()
least_crimes_month = total_crimes_per_month.idxmin()
print(f"The month with the most crimes is {most_crimes_month} with {total_crimes_per_month.max()} crimes.")
print(f"The month with the least crimes is {least_crimes_month} with {total_crimes_per_month.min()} crimes.")
crime_types_per_month = df.groupby(['Month', 'Primary Type']).size().unstack(fill_value=0)
individual_trends = crime_types_per_month.diff().dropna()
for crime_type in crime_types_per_month.columns:
    if crime_types_per_month[crime_type].idxmax() != most_crimes_month or crime_types_per_month[crime_type].idxmin() != least_crimes_month:
        print(f"The crime '{crime_type}' does not follow the overall monthly pattern.")
        print(f"Highest month for '{crime_type}': {crime_types_per_month[crime_type].idxmax()} with {crime_types_per_month[crime_type].max()} crimes.")
        print(f"Lowest month for '{crime_type}': {crime_types_per_month[crime_type].idxmin()} with {crime_types_per_month[crime_type].min()} crimes.")
```
